Importing the necessary libraries:

```python
import numpy as np
import numpy as np
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Input, Concatenate
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.utils import plot_model
```

a)

```python
# Define XOR dataset
x = np.array([[0,0], [0,1],[1,0],[1,1]],"float32")
y = np.array([[0],[1],[1],[0]], "float32")
```

###Without momentum:

```python
#Define model architecture (without momentum)

model = Sequential()
model.add(Dense(2,activation='sigmoid')) #the number of neurons in the layer is 2
model.add(Dense(1,activation='sigmoid')) #the number of neurons in the layer is 1
#Define optimizer and loss function without momentum
model.compile(loss='mean_squared_error', optimizer=SGD(learning_rate=0.1), metrics=['accuracy']) #compute the output and the truth label, for the learning rate, values as 0.1 or 0.2 are recomended

#training the model with the fit method

model.fit(x,y,epochs=1000, batch_size=1)

#to evaluate model

print(model.evaluate(x,y)) #the acccuracy of the model

#predictions:
print(model.predict((x)>0.5).astype('int32')) #if >0.5 should be classe 1, if < 0.5 is classe 0, since it is a sigmoid model (output values are either 0 or 1)
```

```python
model.summary() #for complex models, the model.summary() gives the parameters such as the weight and the bias associated to the network
plot_model(model, show_shapes=True)
```

Here, we can see that the model has two layers: a dense layer with an output shape of (None, 2) and a dense layer with an output shape of (None, 1). The number of parameters associated with each layer is indicated under the "Param #" column. The first layer has 6 parameters (weights and biases), and the second layer has 3 parameters.

###With momentum


```python
#Define model architecture for  with momentum

model2 = Sequential()
model2.add(Dense(2,activation='sigmoid')) #the number of neurons in the layer is 2
model2.add(Dense(1,activation='sigmoid')) #the number of neurons in the layer is 1
#Define optimizer and loss function with momentum
model2.compile(loss='mean_squared_error', optimizer=SGD(learning_rate=0.1, momentum=0.9), metrics=['accuracy'])
#The momentum hyperparameter determines how much of the previous gradients should be used to update the weights of the neural network.
#A value of 0 means that only the current gradient is used to update the weights, while a value of 1 means that only the previous gradient
#is used.

#A common value for momentum is 0.9, which means that the update to the weights is a weighted average of the current gradient
#and the previous update with a weight of 0.9 and 0.1, respectively. This is a good value for many optimization problems, as it allows
#for faster convergence and smoother weight updates. However, the optimal value of momentum can vary depending on the specific problem and dataset.

#training the model with the fit method

model2.fit(x,y,epochs=1000, batch_size=1)

#to evaluate model

print(model2.evaluate(x,y)) #the acccuracy of the model

#predictions:
print(model2.predict((x)>0.5).astype('int32')) #if >0.5 should be classe 1, if < 0.5 is classe 0, since it is a sigmoid model (output values are either 0 or 1)
```

```python
model2.summary()
plot_model(model2, show_shapes=True)
```
